```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn import metrics
```

```python
datab=pd.read_excel(r'C:\Users\Lab4PC17\Desktop\New folder\bill_authentication.xlsx')
datab.head()
```

```python
X=datab[['Variance','Skewness','Curtosis','Entropy']].values
Y=datab['Class'].values
Xtr,Xte,Ytr,Yte=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
svcc=SVC(kernel='linear')
svcc.fit(Xtr,Ytr)
```

```python
Ypr=svcc.predict(Xte)
df = pd.DataFrame({'Actual': Yte.flatten(), 'Predicted': Ypr.flatten()})
df
```

```python
print(metrics.accuracy_score(Yte,Ypr))
print(metrics.precision_score(Yte,Ypr))
print(metrics.recall_score(Yte,Ypr))
```
